## Analysis of the Human Freedom Index

The Human Freedom Index provides a way to understand the relative freedom of countries across the world. It uses specific measures such as government consumption, women safety, and freedom of movement to determine one's economic, civil, and personal freedom in a country, which is then combined to create an overall human freedom score per country.

This analysis focuses on exploring the 2020 Human Freedom Index dataset and looking at how different factors that interest me relate to one another. For example, if there is a relationship between the amount of government consumption and the reliability of the police. Next steps include looking at how different variables may have unexpected interations and how they impact overall human freedom.

The dataset can be found on Kaggle but is originally from a report co-published by the Cato Institute and the Fraser Institute. All features were scored from 0 to 10. https://www.kaggle.com/datasets/gsutters/the-human-freedom-index

### Loading and Filtering the Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
```

```python
# loading the dataset
raw_data = pd.read_csv("hfi_cc_2020.csv")
raw_data.head()
```

```python
# cleaning
# drop rows where hf_score (human freedom score) is empty/NaN
data = raw_data.dropna(subset=['hf_score'])
data.head()
```

```python
# looking at 2018 data
data_2018 = data.loc[data['year'] == 2018]
data_2018.head()
```

```python
# look at USA
data_US = data.loc[data['countries'] == 'United States']
data_US.head()
```

### Visual Exploration

Let's start by looking at the human freedom score across regions.

```python
# group by region and get average human freedom score, plot
grouped_region = data_2018.groupby(data_2018.region)
avg_hf = []
regions = []

for region, df in grouped_region:
    # add mean hf for this region
    avg_hf.append(df['hf_score'].mean())
    regions.append(region)
    
print(regions)
print(avg_hf)
```

```python
ax = sns.barplot(x=avg_hf, y=regions)
ax.set_title("Human Freedom Score by Region")
ax.set_xlabel("Human Freedom Score")
ax.set_ylabel("Region")
```

Let's take a look at the United States in particular over time.

```python
ax = sns.lineplot(x='year',y='hf_score', data=data_US)
ax.set_title('USA Human Freedom Score over Time')
ax.set_xlabel('Year')
ax.set_ylabel('Human Freedom Score')
```

### Relationships between Features

Let's look to see if there's a significant relationship between police reliability and government consumption. Reliability of police means how dependable the police are in terms of enforcing law and order. Government consumption is the percentage of government consumption spending out of total consumption.

```python
# remove NaNs to ensure regression works
data_2018 = data_2018.dropna(subset=['ef_legal_police'])
data_2018 = data_2018.dropna(subset=['ef_government_consumption'])

Y = data_2018['ef_legal_police']
X = data_2018['ef_government_consumption']
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
print("Intercept and slope\n", results.params)
print(results.summary())
print("\np-values\n", results.pvalues)
```

It appears there is a significant linear relationship between government consumption and police reliability (p-value < 0.05).

```python
# graph the relationship
ax = sns.scatterplot(x='ef_government_consumption', y='ef_legal_police', data=data_2018)
```

What about police reliability and personal expression? A higher personal expression score meant individuals were allowed to express themselves freely and in multiple ways.

```python
# remove NaNs to ensure regression works
data_2018 = data_2018.dropna(subset=['ef_legal_police'])
data_2018 = data_2018.dropna(subset=['pf_expression'])

Y = data_2018['pf_expression']
X = data_2018['ef_legal_police']
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
print("Intercept and slope\n", results.params)
print(results.summary())
print("\np-values\n", results.pvalues)
```

The linear regression with personal expression as the dependent variable and police reliability as the independent variable did not show a significant relationship (p = 0.226).

```python
# graph the relationship
ax = sns.scatterplot(x='ef_legal_police', y='pf_expression', data=data_2018)
```
